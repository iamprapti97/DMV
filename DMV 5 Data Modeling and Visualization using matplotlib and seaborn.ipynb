```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('AQI Data Set.csv')
data.head()
```

```python
data.columns
```

```python
column_names=['Id','Months','PM10','SO2','NOx',
             'PM25','NH3','O3','CO','Benzene','AQI']
data.columns=column_names
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
plt.figure(figsize=(15,6))
plt.plot(data['Months'], data['AQI'])
plt.xlabel('Date')
plt.ylabel('AQI')
plt.xticks(rotation=75)
plt.title('Air Quality Index (AQI) Trend Over Time')
plt.show()
```

```python
data.columns
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Months'], data['PM10'], color='red')
plt.xlabel('Date')
plt.xticks(rotation=75)
plt.ylabel('PM10')
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Months'],data['PM25'],color='blue')
plt.xlabel('Date')
plt.xticks(rotation=75)
plt.ylabel('PM25')
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Months'],data['CO'],label='CO',color='orange')
plt.xlabel('Date')
plt.xticks(rotation=75)
plt.ylabel('Carbon Monoxide Levels')
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Months'],data['SO2'],label='SO2',color='violet')
plt.xlabel('Date')
plt.ylabel('SO2')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Months'],data['NOx'],label='NOx',color='purple')
plt.xlabel('Date')
plt.ylabel('NOx')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(data['Months'],data['AQI'],color='skyblue',label='AQI')
plt.xlabel('Months')
plt.ylabel('AQI Value')
plt.legend()
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12,6))
data.boxplot(color='blue')
plt.title('Distribution of AQI Values for Pollutants')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data['AQI'],data['PM25'],c='red',label='PM25')
plt.scatter(data['AQI'],data['SO2'],c='blue',label='SO2')
plt.scatter(data['AQI'],data['O3'],c='green',label='O3')
plt.title('Scatter Plot: AQI vs Pollutant Levels')
plt.xlabel('AQI Value')
plt.ylabel('Pollutant Level')
plt.legend()
plt.grid(True)
plt.show()
```
